```python
lines = ["Hello People\n", "Welcome to Python class\n", "How are you today\n"]
with open("class.txt", "w") as file:
    file.writelines(lines)

with open("class.txt", "r") as files:
    data = files.read()
    print(data)
```

```python
lines.insert(0, "You are awesome\n")
with open("class.txt", "w") as file:
    file.writelines(lines)

with open("class.txt", "r") as files:
    data = files.read()
    print(data)
```

```python
with open("class.txt", "r") as files:
    line1 = files.readline()
    line2 = files.readline()
    print(line1)
    print(line2)
```

```python
with open("class.txt", "r") as files:
    lines = files.readlines()
    print(lines[2])
    print(lines[4])
```
