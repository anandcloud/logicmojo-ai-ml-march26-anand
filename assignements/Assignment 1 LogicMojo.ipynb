Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200,3000,250,7800,540,3000,9100,250,1500,7800]

# Identify and display all transaction values that appear more than once.
duptransactions = []
for tran in transactions:
    if transactions.count(tran) > 1:
        duptransactions.append(tran)
lduptran = list(set(duptransactions))
print(lduptran)    
# Detect transactions greater than ₹5000 as these may require fraud monitoring.
risk_transactions = []
for tran in transactions:
    if tran > 5000:
        risk_transactions.append(tran)
print(risk_transactions)

# Calculate the total spending for the day.
total = 0
for tran in transactions:
    total += tran
print(total)
# Create a new list containing only unique transaction values.
unique_transactions = []
for tran in transactions:
    if tran not in unique_transactions:
        unique_transactions.append(tran)
print(unique_transactions)
        
# Sort the unique transaction list in descending order to view the largest transactions first.
transactions.sort(reverse=True)
print(transactions)
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_ages = []
invalid_counter = 0

# Iterates through the dataset and removes all invalid age values.
for age in ages:
    if age is not None and age > 0 :
        valid_ages.append(age)
    else:
        invalid_counter += 1
# Stores only valid ages in a new list.
print(f"Valid ages list: {valid_ages}")
# Calculates the average age of valid customers.
sum = 0
for vage in valid_ages:
    sum += vage
average_age = sum/len(valid_ages)
print(f"Average age: {average_age}")

# Displays the list of customers whose age is greater than 30.
old_ages = []
for vage in valid_ages:
    if vage > 30:
        old_ages.append(vage)
print(f"Ages greater than 30: {old_ages}")
        
# Prints how many invalid entries were removed during the cleaning process.
print(f"Invalid entries removed during cleaning process: {invalid_counter}")
```

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000] 
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

# Combines the product names and their prices into a single data structure representing a table.
for product, amazon_price, flipkart_price in zip(products, amazon_prices, flipkart_prices):
   # print(f"Product: {product}, Amazon Price ${amazon_price}, Flipkart price ${flipkart_price}")
    price_diff = abs(amazon_price - flipkart_price)
    if amazon_price > flipkart_price:
        print(f"The {product} is ${price_diff} cheaper on Flipkart compared to Amazon.")
    else:
        print(f"The {product} is ${price_diff} cheaper on Amazon compared to Flipkart.")
        
# Compares the price of each product across both platforms.
# Determines which platform offers the lower price.
# Calculates the difference between the two prices.
# Displays the product name, cheaper platform, and price difference.
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
odd_hours_logins = []
even_hours_logins = []
peak_activity_time = []
for login in login_hours:
    if login % 2 == 0:
        even_hours_logins.append(login)
    else:
        odd_hours_logins.append(login)
    if login > 5:
        peak_activity_time.append(login)

print("Odd login hours: " + str(odd_hours_logins))
print("Even login hours: " + str(even_hours_logins))
print("Peak login hours: " + str(peak_activity_time))
sort_hours = sorted(login_hours)
print(sort_hours)
print(len(peak_activity_time))
        
# Separates login hours into even-hour logins and odd-hour logins.
# Identifies login hours that fall into peak activity time (hours greater than 5).
# Sorts the login hours in ascending order.
# Calculates how many logins occurred during peak hours.
```
